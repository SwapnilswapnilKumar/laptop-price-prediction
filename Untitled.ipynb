```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('laptop_price.csv',encoding='latin')
df.head()
```

```python
df=df.drop("Product",axis=1)
```

```python
df["Gpu"] =  df["Gpu"].str.split().str[0]
```

```python
df["Cpu"]
```

```python
cpu_string=[]
for i in df["Cpu"].str.split().str[:2]:
    cpu_string.append(' '.join(i))

df["Cpu"] = pd.DataFrame(cpu_string)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
df["Memory_size"] = df["Memory"].str.split().str.get(0)
df["Memory_type"] = df["Memory"].str.split().str.get(1)
df
```

```python
df["Memory_size"].unique()
```

```python
ls=[]
for i in df["Memory_size"]:
    if "GB" in i:
        ls.append(int(i.replace("GB","")))
    else:
        ls.append(float(i.replace("TB",""))*float(1000))
df["Memory_size"]=ls
df["Memory_size"] = df["Memory_size"].astype(int)
```

```python
df
```

```python
df["Ram"]    = df["Ram"].str.split('G').str.get(0).astype(int)
df["Inches"] = df["Inches"].astype(int)
df["Weight"] = df["Weight"].str.split('k').str.get(0).astype(float)
```

```python
df.head()
```

```python
df["Price_rupee"] = df["Price_euros"]*90.50
```

```python
df.head()
```

```python
df["x_res"] = df["ScreenResolution"].str.split().str[-1].str.split('x').str[0].astype(int)
df["y_res" ] = df["ScreenResolution"].str.split().str[-1].str.split('x').str[1].astype(int)
```

```python
df.head()
```

```python
df["ScreenResolution"] =df["x_res"]*df["y_res"].astype(int)
df
```

```python
df.drop(["Price_euros","Memory","x_res","y_res"],axis=1,inplace=True)
```

```python
df.drop("laptop_ID",axis=1,inplace=True)
```

```python
df.describe()
```

```python
sns.boxplot(df["Price_rupee"])
```

```python
len(df[df["Price_rupee"]>2.6e5]) 
```

```python
df = df[df["Price_rupee"]<2.6e5].reset_index(drop=True)
sns.boxplot(df["Price_rupee"])
```

```python
df.columns
```

```python
df.head()
```

```python
x=df.loc[:,['Company', 'TypeName', 'Inches', 'ScreenResolution', 'Cpu', 'Ram',
       'Gpu', 'OpSys', 'Weight', 'Memory_size', 'Memory_type']]
y=df.loc[:,["Price_rupee"]]
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe.fit(x[['Company', 'TypeName','Cpu','Gpu','OpSys','Memory_type']])
temp = ohe.transform(x[['Company', 'TypeName','Cpu','Gpu','OpSys','Memory_type']]).toarray()
```

```python
ohe.fit_transform(x[['Company', 'TypeName','Cpu','Gpu','OpSys','Memory_type']]).toarray().shape
```

```python
encoded= np.hstack((df[["Inches","ScreenResolution","Ram","Weight","Memory_size","Price_rupee"]].values, temp))
```

```python
encoded
```

```python
encoded_df = pd.DataFrame(encoded)
encoded_df.head()
```

```python
y_new = encoded_df[5]
x_new = encoded_df.drop(5,axis=1)
```

```python
# from sklearn.compose import make_column_transformer
# column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Company', 'TypeName','Cpu','Gpu','OpSys','Memory_type']),remainder="passthrough")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_new,y_new,test_size=0.2)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
linear_obj = LinearRegression()
linear_obj.fit(x_train,y_train)
```

```python
y_pred = linear_obj.predict(x_test)
print("training score of linear regression: ",linear_obj.score(x_train,y_train))
print("testing score of linear regression: ",linear_obj.score(x_test,y_test))
```

### decision Tree

```python
from sklearn import tree
DTrgf = tree.DecisionTreeRegressor()
DTrgf.fit(x_train,y_train)
print("training score of random forest : ",DTrgf.score(x_train,y_train))
print("testing score of random forest : ",DTrgf.score(x_test,y_test))
```

### random forest

```python
from sklearn.ensemble import RandomForestRegressor
RFrgf=RandomForestRegressor(max_depth=7,random_state=1)
RFrgf.fit(x_train,y_train)
print("training score of random forest : ",RFrgf.score(x_train,y_train))
print("testing score of random forest : ",RFrgf.score(x_test,y_test))
```

```python
joblib.dump(linear_obj,'./models/linearRegression.lb')
joblib.dump(DTrgf,'./models/decisionTree.lb')
joblib.dump(RFrgf,'./models/RandomForest.lb')
joblib.dump(ohe,'./models/oneHotEncoder.lb')
```
